```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
scores = pd.read_csv('/home/marthe/Documents/DS/projet/local/frogsdays/scores_v1/recap_scores.csv')
scores = scores.rename(columns = {'f1': 'f1_score'})
```

```python
scores.head()
```

```python
def sup_init_space(element):
    if element[0] == ' ':
        element = element[1:]
    return element

for col in scores.columns[:4]:
    scores[col] = scores[col].apply(lambda x : sup_init_space(x))
```

```python
color_model_dict = {
    'rapidfuzz' : 0,
    'cosine' : 1,
    'knn' : 2,
    'NB' : 3,
    'blastn' : 4 
}

scores = scores.loc[scores['taxo_level'] != 'family']
scores['color'] = scores['model'].apply(lambda x : px.colors.qualitative.G10[color_model_dict[x]])
```

```python
selected_scores = scores.loc[(
    ((scores['ref db'] == 'refseq_in_house') |
    (scores['ref db'] == 'urinary_max10_min2') |
    (scores['ref db'] == 'urinary_max20')) 
    )]
genus_scores = selected_scores.loc[selected_scores['taxo_level'] == 'genus'].reset_index()
species_scores = selected_scores.loc[selected_scores['taxo_level'] == 'species'].reset_index()

genus_scores_dna_seq = genus_scores.loc[genus_scores['region'] == 'dna_seq'].reset_index()
genus_scores_V3V4 = genus_scores.loc[genus_scores['region'] == 'V3V4'].reset_index()

species_scores_dna_seq = species_scores.loc[species_scores['region'] == 'dna_seq'].reset_index()
species_scores_V3V4 = species_scores.loc[species_scores['region'] == 'V3V4'].reset_index()
```

```python
plt.Figure()


fig = px.scatter(
    genus_scores,
    y = 'ref db', 
    x = 'f1_score', 
    color_discrete_sequence=px.colors.qualitative.G10[0:],
    color = 'model', 
    facet_row = 'region', 
    opacity = 0.7)


fig.update_traces(marker = dict(size = 15, line = dict(width = 2, color = 'black')), y0 = 0.5, dy = 0.1)
#fig.update_layout(legend = dict(font = dict(size = 15)))
fig.update_layout(font_size = 14, autosize = False, width = 800, height = 400, showlegend = True, title = dict(text = 'Genus', x = 0.5))
fig.layout.annotations[0]['text'] = 'V3V4'
fig.layout.annotations[1]['text'] = 'complete 16S'
fig.update_annotations(font_size = 16)
fig.update_yaxes(title = None)




fig.show() 
```

```python
plt.Figure()


fig = px.scatter(
    species_scores,
    y = 'ref db', 
    x = 'f1_score', 
    color_discrete_sequence=px.colors.qualitative.G10[1:],
    color = 'model', 
    facet_row = 'region', 
    opacity = 0.7)

fig.update_traces(marker = dict(size = 15, line = dict(width = 2, color = 'black')), y0 = 0.5, dy = 0.1)
#fig.update_layout(legend = dict(font = dict(size = 15)))
fig.update_layout(font_size = 14, autosize = False, width = 800, height = 400, showlegend = True, title = dict(text = 'Species', x = 0.5))
fig.layout.annotations[0]['text'] = 'V3V4'
fig.layout.annotations[1]['text'] = 'complete 16S'
fig.update_annotations(font_size = 16)
fig.update_yaxes(title = None)




fig.show() 
```

```python
plt.Figure()

fig = make_subplots(rows = 2, cols = 1, subplot_titles = ('Genus', 'Species'))

fig.add_trace(go.Scatter(
    mode = 'markers',
    y = genus_scores.loc[genus_scores['region'] == 'dna_seq']['ref db'],
    x = genus_scores.loc[genus_scores['region'] == 'dna_seq']['f1_score'],
    marker = dict(
        color = 'grey',
         size = 50),
    #y = 'ref db', 
    #x = 'f1_score', 
    #color_discrete_sequence=px.colors.qualitative.G10[1:],
    opacity = 0.7,
    name = 'complete 16S'
    ))

fig.add_trace(go.Scatter(
    mode = 'markers',
    y = genus_scores.loc[genus_scores['region'] == 'V3V4']['ref db'],
    x = genus_scores.loc[genus_scores['region'] == 'V3V4']['f1_score'],
    marker = dict(
        color = 'grey',
        symbol = 22,
        size = 50,
        ),
    #y = 'ref db', 
    #x = 'f1_score', 
    #color_discrete_sequence=px.colors.qualitative.G10[1:],
    opacity = 0.7,
    name = 'V3-V4'))

fig.update_layout(xaxis_range = [0,1])
fig.update_legends(font_size = 20, orientation = 'h', title = 'Region')
fig.update_traces(marker = dict(size = 30, line = dict(width = 3, color = 'black')))
```

```python

genus_traces = [
    go.Scatter(
    mode = 'markers',
    y = genus_scores_dna_seq['ref db'],
    x = genus_scores_dna_seq['f1_score'],
    marker = dict(
        color = genus_scores_dna_seq['color'], 
        size = 60,
        line = dict(width =1, color = 'black')
        )
    ),

    go.Scatter(
    mode = 'markers',
    y = genus_scores_V3V4['ref db'],
    x = genus_scores_V3V4['f1_score'],
    marker = dict(
        color = genus_scores_V3V4['color'],
        symbol = 22,
        size = 30,
        line = dict(width =2, color = 'black')
        ))
]


species_traces = [
    go.Scatter(
    mode = 'markers',
    y = species_scores_dna_seq['ref db'],
    x = species_scores_dna_seq['f1_score'],
    marker = dict(
        color = species_scores_dna_seq['color'],
        size = 60,
        line = dict(width =1, color = 'black')
        ),
    opacity = 0.7
    ),
    
    go.Scatter(
    mode = 'markers',
    y = species_scores_V3V4['ref db'],
    x = species_scores_V3V4['f1_score'],
    marker = dict(
        color = species_scores_V3V4['color'],
        symbol = 22,
        size = 30,
        line = dict(width =2, color = 'black')
        ),
    opacity = 0.8)
]




fig = make_subplots(rows = 2, cols = 1)

for trace in genus_traces:
    fig.add_trace(trace, row = 1, col = 1)
    
for trace in species_traces:
    fig.add_trace(trace, row = 2, col = 1)
    
fig.update_xaxes(range = [0,1], dtick = 0.1)
fig.update_yaxes(ticktext = ['refseq_in_house', 'urinary_2_10', 'urinary_1_20'], tickvals = ['0', '1', '2'])
fig.update_layout(font_size = 50, autosize = False, width = 6000, height = 1200, showlegend = False)
fig.update_annotations(font_size = 70)
fig.show()



```
