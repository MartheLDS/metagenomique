```python
import pandas as pd
```

```python
db = 'urinary_max10_min2'
region = 'dna_seq'
file = '../datasets/train_sets/'+db+'.csv'
output = '../datasets/train_sets/'+db+'_'+region+'.fasta'
```

```python
data = pd.read_csv(file)
```

```python
data.shape
```

```python
data
```

```python
def rm_spaces(seq_id):
    seq_id = seq_id.replace(' ', '_')
    return seq_id

data['seq_id'] = data['seq_id'].apply(lambda x : rm_spaces(x))
data
```

```python
with open(output, mode = 'a') as file:
    for k in range(data.shape[0]):
        test = data['seq_id'][k] + '\n' + data['V3V4'][k] + '\n'
        file.write(test)
```
